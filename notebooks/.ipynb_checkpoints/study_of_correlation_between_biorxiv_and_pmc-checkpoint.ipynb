### Time Lagged Cross Correlation — assessing signal dynamics

```python
def crosscorr(datax, datay, lag=0, wrap=False):
    """ Lag-N cross correlation. 
    Shifted data filled with NaNs 
    
    Parameters
    ----------
    lag : int, default 0
    datax, datay : pandas.Series objects of equal length
    Returns
    ----------
    crosscorr : float
    """
    if wrap:
        shiftedy = datay.shift(lag)
        shiftedy.iloc[:lag] = datay.iloc[-lag:].values
        r = datax.corr(shiftedy)
    else: 
        print(datay.shift(lag))
        r = datax.corr(datay.shift(lag))
    
    if not np.isnan(r):
        return r
    else:
        return 0.0

avg_rs = np.zeros(24)
for phrase in random_phrases:
    pmc_name = "pmc_"+phrase
    biorxiv_name = "biorxiv_"+phrase
    d1 = df[pmc_name]
    d2 = df[biorxiv_name]

    rs = [crosscorr(d1,d2, lag) for lag in range(-12,12)]
    avg_rs += np.array(rs)

avg_rs = avg_rs/len(random_phrases) 
offset = np.ceil(len(avg_rs)/2)-np.argmax(avg_rs)
f,ax=plt.subplots(figsize=(14,3))
ax.plot(avg_rs)
ax.axvline(np.ceil(len(avg_rs)/2),color='k',linestyle='--',label='Center')
ax.axvline(np.argmax(avg_rs),color='r',linestyle='--',label='Peak synchrony')
ax.set(title=f'Offset = {offset} frames\nPMC leads <> biorxiv leads', xlabel='Offset',ylabel='Pearson r')
ax.set_xticks([0, 4, 8, 12, 16, 20, 24])
ax.set_xticklabels([-12, -8, -4, 0, 4, 8, 12]);
plt.legend()
```

```python
def crosscorr(datax, datay, lag=0, wrap=False):
    """ Lag-N cross correlation. 
    Shifted data filled with NaNs 
    
    Parameters
    ----------
    lag : int, default 0
    datax, datay : pandas.Series objects of equal length
    Returns
    ----------
    crosscorr : float
    """
    if wrap:
        shiftedy = datay.shift(lag)
        shiftedy.iloc[:lag] = datay.iloc[-lag:].values
        r = datax.corr(shiftedy)
    else: 
        r = datax.corr(datay.shift(lag))
    
    if not np.isnan(r):
        return r
    else:
        return 0.0

no_splits = 3
samples_per_split = df.shape[0]/no_splits
rss=[]
for t in range(0, no_splits):
    avg_rs = np.zeros(24)
    for phrase in random_phrases:
        pmc_name = "pmc_"+phrase
        biorxiv_name = "biorxiv_"+phrase
        d1 = df[pmc_name].loc[(t)*samples_per_split:(t+1)*samples_per_split]
        d2 = df[biorxiv_name].loc[(t)*samples_per_split:(t+1)*samples_per_split]

        rs = [crosscorr(d1,d2, lag) for lag in range(-12,12)]
        avg_rs += np.array(rs)
    
    avg_rs = avg_rs/len(random_phrases)
    rss.append(avg_rs)
    
rss = pd.DataFrame(rss)
f,ax = plt.subplots(figsize=(10,5))
sns.heatmap(rss,cmap='RdBu_r',ax=ax)
ax.set(title=f'Windowed Time Lagged Cross Correlation', xlabel='Offset',ylabel='Window epochs')
```

```python
# Rolling window time lagged cross correlation
window_size = 24 #samples
t_start = 0
t_end = t_start + window_size
step_size = 1
rss=[]
while t_end < 72:
    avg_rs = np.zeros(24)
    for phrase in random_phrases:
        pmc_name = "pmc_"+phrase
        biorxiv_name = "biorxiv_"+phrase
        d1 = df[pmc_name].iloc[t_start:t_end]
        d2 = df[biorxiv_name].iloc[t_start:t_end]
        rs = [crosscorr(d1,d2, lag, wrap=False) for lag in range(-12,12)]
        avg_rs += np.array(rs)
    avg_rs = avg_rs/len(random_phrases)
    rss.append(avg_rs)
    t_start = t_start + step_size
    t_end = t_end + step_size
rss = pd.DataFrame(rss)

f,ax = plt.subplots(figsize=(10,10))
sns.heatmap(rss,cmap='RdBu_r',ax=ax)
ax.set(title=f'Rolling Windowed Time Lagged Cross Correlation', xlabel='Offset',ylabel='Epochs')
```

```python
# Rolling window time lagged cross correlation
window_size = 24 #samples
t_start = 0
t_end = t_start + window_size
step_size = 1
rss=[]
while t_end < 72:
    avg_rs = np.zeros(24)
    for phrase in random_phrases:
        pmc_name = "pmc_"+phrase
        biorxiv_name = "biorxiv_"+phrase
        d1 = df[pmc_name].iloc[t_start:t_end]
        d2 = df[biorxiv_name].iloc[t_start:t_end]
        rs = [crosscorr(d1,d2, lag, wrap=False) for lag in range(-12,12)]
        avg_rs += np.array(rs)
    avg_rs = avg_rs/len(random_phrases)
    rss.append(avg_rs)
    t_start = t_start + step_size
    t_end = t_end + step_size
rss = pd.DataFrame(rss)

f,ax = plt.subplots(figsize=(10,10))
sns.heatmap(rss,cmap='RdBu_r',ax=ax)
ax.set(title=f'Rolling Windowed Time Lagged Cross Correlation', xlabel='Offset',ylabel='Epochs')
```
