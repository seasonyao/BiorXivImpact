```python
import pandas as pd

df  = pd.read_pickle("Path/to/pickle")

def get_year(x):
    return int(x['date'].split(',')[1])

df['year'] = df.apply(get_year, axis=1)

df.head()
```

```python
from collections import defaultdict

global doc_occurrence_dict #Number of documents the phrase occured in
global score_dict #Average score for each phrase

phrases_dict = defaultdict(list)

def cal_docOccurNum_freq_score(x):
    global doc_occurrence_dict
    global score_dict
    for phrase in x['key_phrases_for_abstract']:
        doc_occurrence_dict[phrase[2]] += 1
        score_dict[phrase[2]] += phrase[0]

for year in df['year'].value_counts().index.tolist(): 
    doc_occurrence_dict = defaultdict(int)
    score_dict = defaultdict(int)
    
    df[df['year']==year].apply(cal_docOccurNum_freq_score, axis=1)
    
    for phrase in list(doc_occurrence_dict.keys()):
        phrases_dict[phrase].append([year, doc_occurrence_dict[phrase], score_dict[phrase]/doc_occurrence_dict[phrase]])
```

```python
dict_for_df = {}
years = df['year'].value_counts().index.tolist()
years.sort()

for phrase in list(phrases_dict.keys()):
    tmp_freq = []
    total_occur_docs_num = 0
    total_scores = 0
    for year in years:
        for item in phrases_dict[phrase]:
            if item[0] == year:
                total_occur_docs_num+=item[1]
                total_scores+=item[2]
                flag = 0
                break
    
    tmp_freq.append(total_scores)
    tmp_freq.append(total_occur_docs_num)
    
    dict_for_df[phrase] = tmp_freq
```

```python
phrases_name = []
occur_docs_num = []
scores = []

for phrase in list(dict_for_df.keys()):
    phrases_name.append(phrase)
    occur_docs_num.append(dict_for_df[phrase][-1])
    scores.append(dict_for_df[phrase][-2]/dict_for_df[phrase][-1])
    
df_with_occur_doc_num = pd.DataFrame({'phrases_name':phrases_name, 
                   'occur_docs_num':occur_docs_num,  
                   'scores':scores})
df_with_occur_doc_num
```

```python
df_with_occur_doc_num_more_than_1 = df_with_occur_doc_num[df_with_occur_doc_num['occur_docs_num']!=1]
df_with_occur_doc_num_more_than_1
```

```python
df_with_occur_doc_num_more_than_1 = df_with_occur_doc_num_more_than_1.sort_values(by=['scores'])
df_with_occur_doc_num_more_than_1
```

```python
df_with_occur_doc_num_more_than_1.to_pickle("path/to/output/pickle")
```
